### Calculate the Average of Selected Column

```python
Avg_price=df["price"].mean()
```

```python
print("Average House Price",Avg_price)
```

```python
Avg_bedsRooms=df["bedrooms"].mean()
```

```python
print("Average Bedrooms",Avg_bedsRooms)
```

```python
TopCities=df.groupby("city")["price"].mean().sort_values(ascending=False)
```

```python
print("Most Expensive Cities: \n",TopCities.head(10))
```

```python
topcityprices = df.groupby("city")["price"].mean().sort_values(ascending=False).head(10)
```

```python
plt.figure(figsize=(12,6))
topcityprices.plot(kind='bar', color='skyblue')
plt.title("Top 10 Cities by Average House Price")
plt.ylabel("Average Price")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x='sqft_living', y='price', data=df, alpha=0.6)
plt.title("Living Area vs. House Price")
plt.xlabel("Square Feet of Living Area")
plt.ylabel("Price")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(14,10))
numeric_df = df.select_dtypes(include=['int64', 'float64'])
corr = numeric_df.corr()
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```
